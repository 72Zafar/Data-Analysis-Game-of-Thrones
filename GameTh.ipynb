Game of Thrones Battle Analysis This project explores and analyzes data from battles in the "Game of Thrones" universe, specifically from the War of the Five Kings. The dataset battles.csv contains detailed information about various battles, including the participating kings, locations, army sizes, outcomes, and commanders. Additionally, the project incorporates data from the character-deaths.csv file to provide insights into the deaths of characters throughout the series.

Data Overview Battles Data (battles.csv):

Contains 38 rows and 25 columns. Key columns include: name: Name of the battle. year: Year the battle occurred. attacker_king and defender_king: The kings leading the attack and defense. attacker_size and defender_size: The sizes of the attacking and defending armies. battle_type: Type of battle (e.g., pitched battle, siege). Many columns have missing values, especially in secondary attackers and defenders, indicating the primary forces involved were often supported by smaller or less significant troops. Character Deaths Data (character-deaths.csv):

Contains 917 rows and 13 columns. Includes data on character deaths, allegiances, gender, nobility, and the book and chapter of their death. Key insights include: The majority of characters who died were male. A significant proportion of deceased characters were of noble birth. Key Findings Kings and Battles:

Joffrey/Tommen Baratheon was the most frequent attacker king, involved in 14 battles, followed by Robb Stark with 10 battles. Locations with the most battles included Riverrun and Winterfell, each hosting multiple significant confrontations. Army Sizes:

Visualizations using Seaborn bar plots highlighted the varying sizes of armies led by different kings. However, missing data for some battles led to warnings when calculating mean values. Major Deaths and Outcomes:

The project analyzed the relationship between kings and major deaths in battles, demonstrating the deadly nature of some leaders' campaigns. Character Deaths:

A bar plot of character deaths by year shows peaks in specific years, aligning with major events in the series. Analysis of allegiances provides insights into which houses suffered the most losses. Visualizations Several visualizations were created to better understand the data, including:

Bar plots of army sizes by attacker and defender kings. Analysis of major deaths by attacker kings. Distribution of character deaths over years and by allegiances. Conclusion This analysis provides a data-driven look at the conflicts in "Game of Thrones," revealing patterns in leadership, battle outcomes, and the fates of characters. It showcases the complexity and brutality of the wars that shaped the series' narrative.
